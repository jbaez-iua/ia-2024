```python
import requests
import json

OLLAMA_API_URL = "http://localhost:11434/api/generate"
```

```python
def consultar_ollama(prompt, modelo="tinyllama"):
    """
    Envía una consulta a la API de Ollama y devuelve la respuesta.
    """
    datos = {
        "model": modelo,
        "prompt": prompt,
        "stream": False
    }
    try:
        respuesta = requests.post(OLLAMA_API_URL, json=datos)
        
        # Imprimir información de depuración
        print(f"Código de estado: {respuesta.status_code}")
        print(f"Encabezados: {respuesta.headers}")
        print(f"Contenido de la respuesta: {respuesta.text}")
        
        # Intentar parsear la respuesta como JSON
        try:
            resultado = respuesta.json()
            print(f"Respuesta JSON parseada: {json.dumps(resultado, indent=2)}")
            
            if isinstance(resultado, dict) and 'response' in resultado:
                return resultado['response']
            else:
                print("La respuesta no tiene el formato esperado.")
                return respuesta.text
        except json.JSONDecodeError:
            print("La respuesta no es JSON válido.")
            return respuesta.text
    
    except requests.RequestException as e:
        print(f"Error en la solicitud: {e}")
        return None

```

```python
pregunta = "¿Cuáles son los principales desafíos del cambio climático?"
    
print("Prompt Básico:")
respuesta = consultar_ollama(pregunta)
print(respuesta)    
```

```python
pregunta = "¿Cuáles son los principales desafíos del cambio climático?"
    
print("Prompt Básico:")
respuesta = consultar_ollama(pregunta, "qwen:0.5b")
print(respuesta)    
```
